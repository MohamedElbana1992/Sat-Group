```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('datasets/Restaurant_Reviews.csv',sep='\t')
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer=CountVectorizer(stop_words='english')

vector=vectorizer.fit_transform(df['Review'])
vector.toarray()
```

```python
df_res=pd.DataFrame(vector.toarray(),columns=vectorizer.get_feature_names())
df_res
```

```python
df['Review'][0]
```

```python
df_res['place'][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer=TfidfVectorizer(stop_words='english')

vector=vectorizer.fit_transform(df['Review'])
vector.toarray()
```

```python
df_res2=pd.DataFrame(vector.toarray(),columns=vectorizer.get_feature_names())
df_res2
```
